```python
from fuzzywuzzy import process

def correct_specialization_spelling_errors(target_word = "", Choices = [], threshold = 80):
    match, score = process.extractOne(target_word, Choices)
    
    if score >= threshold:
        return match
    else:
        return target_word
```

#### Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.” 
- Test this claim with the data given to you.

```python
# Filter the employee with the specialization of Computer Science and Engineering
emplyee_with_specialization_cse = dataset[dataset_dt["Specialization"] == "Computer Science and Engineering"]
# Define the role that you want to check 
required_roles = ["programming analyst", "software engineer", "hardware engineer", "associate engineer"]
# Filter the employee who has pursued their carrer in CSE and working as specific role in required_roles
filtered_data = dataset_dt[dataset_dt["Designation"].isin(required_roles)]

# Define an empty dictionary
average_salaries = {}
for role in required_roles:
    role_salaries = filtered_data[filtered_data["Designation"] == role]["Salary"]
    average_salaries[role] = role_salaries.mean()/100000
```

```python
for role, avg_salary in average_salaries.items():
    print(f"Average salary for {role}: {avg_salary:.2f} lakhs per annum")
```

```python
claimed_salary_range = (2.5, 3)
print("Claimed Salary Range: 2.5 - 3 lakhs per annum")
for role, avg_salary in average_salaries.items():
    if avg_salary >= claimed_salary_range[0] and avg_salary <= claimed_salary_range[1]:
        print(f"\nThe claim holds true for the {role}.")
    else:
        print(f"\nThe claim doesn't hold true for the {role}.")
```

#### Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)

```python
from scipy.stats import chi2_contingency
```

```python
contingency_table = pd.crosstab(dataset_dt["Gender"], dataset_dt["Specialization"])
chi2, p, _, _ = chi2_contingency(contingency_table)

print("Chi-square Test Result:")
print("________________________")
print("\nChi-square Statistic:", round(chi2, ndigits = 2))
print("\np-value:", p)


alpha = 0.05
if p < alpha:
    print("\nWe reject null hypothesis, there is a significant dependency between gender and specialization.")
else:
    print("\nWe fail to reject null hypothesis, there is no significant dependency between gender and specialization.")
```
